```python
import cv2
import mediapipe as mp
import numpy as np

mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils

with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    cap = cv2.VideoCapture("C://Users//Ravikrishna J//Downloads//dance2.mp4")
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break  # Break out of the loop if there are no more frames to read
        
        # Resize frame to fit the window
        frame = cv2.resize(frame, (640, 480))  # Adjust size as needed
        
        # Recolor image to RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
      
        # Make detection
        results = pose.process(image)
    
        # Recolor back to BGR
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        # Render detections
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                   mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=2),
                                   mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2))               
        
        # Extract and display landmark points in a separate frame
        landmarks_frame = np.zeros_like(image)
        if results.pose_landmarks:
            # Draw lines connecting all keypoints with thicker lines
            for connection in mp_pose.POSE_CONNECTIONS:
                start_idx, end_idx = connection
                start_point = tuple(np.multiply([results.pose_landmarks.landmark[start_idx].x, 
                                                 results.pose_landmarks.landmark[start_idx].y], [640, 480]).astype(int))
                end_point = tuple(np.multiply([results.pose_landmarks.landmark[end_idx].x, 
                                               results.pose_landmarks.landmark[end_idx].y], [640, 480]).astype(int))
                cv2.line(landmarks_frame, start_point, end_point, (0, 255, 0), 5)
            
            # Draw the head with a larger circle
            if (results.pose_landmarks.landmark[mp_pose.PoseLandmark.NOSE] and
                results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_EAR] and
                results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_EAR]):
                
                # Calculate the center of the head
                nose_x = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.NOSE].x * image.shape[1])
                nose_y = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.NOSE].y * image.shape[0])
                left_ear_x = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_EAR].x * image.shape[1])
                right_ear_x = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_EAR].x * image.shape[1])
                
                # Calculate the diameter of the head
                head_diameter = int(np.sqrt((right_ear_x - left_ear_x) ** 2 + (nose_y - nose_y) ** 2))
                
                # Draw the head as a circle with thicker line
                cv2.circle(landmarks_frame, (nose_x, nose_y), head_diameter // 2, (0, 255, 0), 10)
            
            # Draw circles for all keypoints
            for landmark in results.pose_landmarks.landmark:
                height, width, _ = image.shape
                cx, cy = int(landmark.x * width), int(landmark.y * height)
                cv2.circle(landmarks_frame, (cx, cy), 8, (0, 255, 0), cv2.FILLED)

        cv2.imshow('Mediapipe Feed', image)
        cv2.imshow('Landmarks', landmarks_frame)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
```
